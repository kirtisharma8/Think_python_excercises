```python
#in text function that counts how many times a letter appear in the string using dict,
def histogram(s):
    d=dict()
    for c in s:
        #below checks if the char is not in the dict then adds else it just increases iits counter
        if c not in d:
            d[c]=1
        else:
            d[c]=d[c]+1
    return d
h=histogram('brontosaurus')
h
```

```python
#get method of dictionaries, it takes key and a default value as a parameter if key does not exist then it returns the default
#value for example below
h=histogram('aa')
h
h.get('a',0)
h.get('b',0)
```

```python
#looping and dict, for loop traverses through the keys of dict
def print_hist(h):
    for c in h:
        print(c,h[c])
h=histogram('parrot')
print_hist(h)
```

```python
#sorted mehtod to sort the keys in dict
for key in sorted(h):
    print(key,h[key])
```

```python
#Reverse lookup
def reverse_lookup(d,v):
    for k in d:
        if d[k]==v:
            return k
    raise LookupError('Value does not appear in the dictionary')
h=histogram('parrot')
k=reverse_lookup(h,2)
k
```

```python
#reverse look up that raise an error
k=reverse_lookup(h,3)
k
```

```python
#dictionaries and lists
#lists can be stored as values in the dict
#below function reveserse the above deict and store letter as lists and frequencies as values
def invert_dict(d):
    inverse=dict()
    for key in d:
        val=d[key]
        if val not in inverse:
            inverse[val]=[key]
        else:
            inverse[val].append(key)
    return inverse
hist=histogram('parrot')
```

```python
print (hist)
```

```python
inverse=invert_dict(hist)
inverse
```

```python
#fibonacci using dictionaries- memoization
known={0:0,1:1}
def fibonacci(n):
    if n in known:
        return known[n]
    res=fibonacci(n-1)+fibonacci(n-2)
    known[n]=res
    return res

fibonacci(34)
```

```python
#ex11.1 write a function that reads the words in word.txt and stores them as keys in a dictionary. It doesn't matter what
#values are. Then you can use in operator as a fast way to check whether a strings in the dictionary

def read_file(file):
    fin=open(file)
    d1 = dict()
    key=0
    for line in fin:
        word=line.strip()
        d1[word]=key
        key=key+1
    return d1

d3=read_file('words.txt')
d3

#below function checks if any string is in the dictionary or not
def strin_check(d,s):
    if s in d:
        return True
    else:
        return False
    
strin_check(d3,'abolaa')
```

```python
#ex11.2- using setdefault method of dictionaries rewrite invert_dict 
def invert_dict2(d):
    inverse=dict()
    for key in d:
        val=d[key]
        #If key is in the dictionary, return its value.
        #If not, insert key with a value of default and return default. default defaults to None.
        inverse.setdefault(val, []).append(key)
    return inverse

hist=histogram('parrot')
print(hist)
inverse2=invert_dict2(hist)
inverse2
```

```python
#Ex11.4 has_duplicates using dictionaries

def has_duplicates(t):
    d={}
    for x in t:
        if x in d:
            return True
        d[x]=True
    return False

t = [1, 2,3]
print(has_duplicates(t))
t.append(1)
print(has_duplicates(t))

        
    
    
```
