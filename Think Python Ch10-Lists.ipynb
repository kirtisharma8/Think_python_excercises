```python
#below function traverse one list while building another
#below function takes a list of strings and returns a new list by capatilizing the strings
def capitalize_all(t):
    res=[]
    for s in t:
        res.append(s.capitalize())
    return res
capitalize_all(['a','g','k'])
```

```python
#below function return sublist by selecting uppercase elements from a list
#operation like this is called filter because it is filtering the list
def only_upper(t):
    res=[]
    for s in t:
        if s.isupper():
            res.append(s)
    return res
only_upper(['a','Ll','bc','K','OP'])
```

```python
#Ex10.1 Write a function that takes a list of lists of intergers and addes up the elements from all of the nested lists.
def nested_sum(t):
    total=0
    for s in t:
        total = sum(s) + total
    return total
nested_sum([[1,2],[3,4],[4,5,6],[5]])
```

```python
#Ex10.2 write a function called cumsum that takes a list of number and returns cumulative sum that is a new list where the ith
#element is the sum of the first i+1 elements form the original list
#cumulative sum is each element is the sum of all elements till that index
def cumsum(t):
    final_list=[]
    total=0
    for i in t:
        total = i + total
        final_list.append(total)
    return final_list

cumsum([1,2,7,4])
    


```

```python
#ex10.3 write a function called middle that takes a list and returns a new list that contains all but not first and last

def middle(t):
    t.pop(0)
    t.pop(len(t)-1)
    return t
middle([1,2,3,4,5])

#another way to do it is using list slice if we return list [1:-1] this will return the list middle starting at index 1 and 
#eleminating 0
def middle2(t):
    return t[1:-1]
middle2([1,2,3,4,5])
```

```python
# EX10.6 Two words are anagrams if you can rearrange the one to spell other, for ex silent can be arranged to spell listen
# below function checks if the word is anagram, logic is both should have same letters so its smart to use list method sort
# to verify that if both are anagrams
def is_anagram(word1,word2):
    return sorted(word1)==sorted(word2)
print(is_anagram('abba', 'baba'))
print(is_anagram('lol', 'kol'))
print(is_anagram('a', 'b'))
print(is_anagram('juio', 'ioju'))
```

```python
#ex10.7 write a function called has_duplicates that takes a lis and returns true if there is any element that appears more than
# once it should not modify the original list
def has_duplicates(t):
    t2=list(t) #copying list so that we do not modify the original
    t2.sort()
    for i in range(len(t2)-1):
        if t2[i]==t2[i+1]:
            return True
    return False
print(has_duplicates(['jjjkj','jjjkj']))
print(has_duplicates(('jjjkj')))
print(has_duplicates([1,3,4,5,6]))
```

```python
#part of assigenment probablility
import math
import sys

def birthday_probability(n, d):
    """Calculate the probability of generating a duplicate random number after
    generating "n" random numbers in the range "d".
    """
    # Formula taken from: https://en.wikipedia.org/wiki/Birthday_problem
    return 1 - math.e ** (-n**2 / (2 * d))
```

```python
#ex10.9 Write a function that reads the file words.txt and builds a list one element per word.
#looked at book solution
def read_file(file):
    fin=open(file)
    l=[]
    for line in fin:
        word=line.strip()
        l.append(word)
    return l

li=read_file('words.txt')
```

```python
#ex10.10 Write a function called in_bisect that takes a sorted list and a target value and returns the index of the 
#value in the list if its there or none if its not
#looked at the book solution not written from scratch, but understood what author has done
def in_bisect(li,word):
    if len(li)==0:
        return False
    i=len(li)//2  #// is floor division so that it provides the next rounded number not decimal
    if li[i]==word:
        return True
    if li[i]>word:
        #it will make a recursive call of the same function by selecting first half right before i and searching first half
        return in_bisect(li[:i], word)
    else:
        #it will search the second half
        return in_bisect(li[i+1:],word)
in_bisect(li,'yashu')
```
